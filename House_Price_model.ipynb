```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from IPython.display import display
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
pd.options.display.max_columns = None
display(train_data)
display(train_data.describe(include='all'))
display(test_data)
display(test_data.describe(include='all'))
```

```python
raw_train_data = train_data.iloc[:,0:-1]
y = train_data['SalePrice']
raw_test_data = test_data.copy()
```

```python
raw_data = pd.concat([raw_train_data,raw_test_data],ignore_index=True,axis=0)
raw_data
```

```python
sns.heatmap(raw_data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
missing_df = pd.DataFrame()
missing_df['nan_features'] = [feature for feature in raw_data.columns if raw_data[feature].isnull().sum()>0]
missing_df['number_of_missing'] = [raw_data[feature].isnull().sum() for feature in missing_df['nan_features']]
missing_df['percent_missing'] = [raw_data[feature].isnull().sum()/raw_data.shape[0] for feature in missing_df['nan_features']]
missing_df['feature_type'] = [np.dtype(raw_data[feature]) for feature in missing_df['nan_features']]
print(missing_df)
```

```python
for feature in ['MiscFeature','PoolQC','Fence','FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2']:
    raw_data[feature].replace(np.nan,'NA',inplace=True)

raw_data['MasVnrType'].replace(np.nan,'None',inplace=True)
raw_data['MasVnrArea'].replace(np.nan,0,inplace=True)
```

```python
for column in ['MSZoning','Utilities','Exterior1st','Exterior2nd','Electrical','KitchenQual','Functional','SaleType']:
    raw_data[column] = raw_data[column].fillna(raw_data[column].mode().iloc[0])

for column in ['BsmtFullBath','BsmtHalfBath','GarageCars']:
    raw_data[column] = raw_data[column].fillna(raw_data[column].median())
    
for column in ['BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','GarageArea']:
    raw_data[column] = raw_data[column].fillna(raw_data[column].mean())
```

```python
def round(n):
    if n%10<5:
        return n-n%10
    elif n%10==5:
        return n
    elif n%10>5:
        return n+(10-n%10)
    else:
        return n

raw_data['LotFrontage'] = raw_data['LotFrontage'].apply(round)

plt.figure(figsize=(20,12))
sns.boxplot(x='LotConfig',y='LotFrontage',hue='LotShape',data=raw_data,palette='winter')
```

```python
def fill_LotFrontage(column):
    LotFrontage = column[0]
    LotConfig = column[1]
    LotShape = column[2]
    
    if pd.isnull(LotFrontage):
        
        if LotConfig=='Inside':
            return 60
        
        elif LotConfig=='FR2':
            if LotShape!='IR2':
                return 60
            else:
                return 50
        
        elif LotConfig=='Corner':
            if LotShape=='Reg':
                return 60
            elif LotShape=='IR3':
                return 200
            else:
                return 90
        
        elif LotConfig=='CulDSac':
            if LotShape=='Reg':
                return 30
            elif LotShape=='IR3':
                return 100
            else:
                return 50
        
        elif LotConfig=='FR3':
            if LotShape!='IR1':
                return 70
            else:
                return 60
        
    else:
        return LotFrontage
    
raw_data['LotFrontage'] = raw_data[['LotFrontage','LotConfig','LotShape']].apply(fill_LotFrontage,axis=1)
```

```python
from scipy import stats
def diagnostic_plots(variable):
    # function to plot a histogram and a Q-Q plot
    # side by side, for a certain variable
    
    plt.figure(figsize=(15,6))
    plt.subplot(1, 2, 1)
    sns.distplot(variable)
    
    plt.subplot(1, 2, 2)
    stats.probplot(variable, dist="norm", plot=plt)

    plt.show()

diagnostic_plots(y)
```

```python
log_y = np.log(y)
```

```python
raw_data1 = pd.get_dummies(raw_data,drop_first=True)
raw_data1
```

```python
X = raw_data1.iloc[0:1460].values
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor_rf = RandomForestRegressor(random_state=0)
```

```python
from sklearn.svm import SVR
regressor_svr = SVR(kernel = 'rbf')
```

```python
import xgboost as xgb
regressor_xgb = xgb.XGBRegressor()
```

```python
from sklearn.model_selection import KFold

n_folds = 10
scores = []
def k_fold(regressor):
    
    cv = KFold(n_splits=10, random_state=42, shuffle=False)
    for train_index, test_index in cv.split(X):
        #print("Train Index: ", train_index, "\n")
        #print("Test Index: ", test_index)

        X_train, X_test, y_train, y_test = X[train_index], X[test_index], log_y[train_index], log_y[test_index]
        regressor.fit(X_train, y_train)
        scores.append(regressor.score(X_test, y_test))
    print(np.mean(scores))
```

```python
k_fold(regressor_rf)
```

```python
k_fold(regressor_svr)
```

```python
k_fold(regressor_xgb)
```
